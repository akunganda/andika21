```python
from pandas.core.arrays.categorical import console
import pandas as pd
import numpy as np
import missingno as msno
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
```

```python
dataset = pd.read_excel('Stunting1.xlsx')
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
df_cleaned = dataset.dropna()
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned
```

```python
# Encoding variabel kategorikal menggunakan LabelEncoder
encoder = LabelEncoder()
df_cleaned = pd.DataFrame(df_cleaned)
df_cleaned['TB/U'] = encoder.fit_transform(df_cleaned['TB/U'])
```

```python
scaler = MinMaxScaler()
df = pd.DataFrame(df_cleaned)
df['BB_Lahir'] = scaler.fit_transform(df[['BB_Lahir']])
df['TB_Lahir'] = scaler.fit_transform(df[['TB_Lahir']])
df['Usia_Saat_Ukur'] = scaler.fit_transform(df[['Usia_Saat_Ukur']])
df['Berat'] = scaler.fit_transform(df[['Berat']])
df['Tinggi'] = scaler.fit_transform(df[['Tinggi']])
df['ZS_BB/U'] = scaler.fit_transform(df[['ZS_BB/U']])
df['ZS_TB/U'] = scaler.fit_transform(df[['ZS_TB/U']])
df['ZS_BB/TB'] = scaler.fit_transform(df[['ZS_BB/TB']])
df['ZS_TB/U'] = scaler.fit_transform(df[['ZS_TB/U']])
```

```python
df.to_csv('dataset.csv', index=False)
```
